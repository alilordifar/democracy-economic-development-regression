```python
trade = trade.drop(columns=["Indicator Name", "Indicator Code"])
inflation = inflation.drop(columns=["Indicator Name", "Indicator Code"])
unemployment = unemployment.drop(columns=["Indicator Name", "Indicator Code"])
urban = urban.drop(columns=["Indicator Name", "Indicator Code"])
```

```python
trade = trade.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="trade"
)

inflation = inflation.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="inflation"
)

unemployment = unemployment.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="unemployment"
)

urban = urban.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="urban"
)
```

```python
def clean_year(df):
    df = df[df["year"].str.isdigit()].copy()   # 👈 important
    df.loc[:, "year"] = df["year"].astype(int)
    return df
```

```python
gdp_per_capita = clean_year(gdp_per_capita)
population = clean_year(population)
life_expectancy = clean_year(life_expectancy)
education = clean_year(education)
trade = clean_year(trade)
inflation = clean_year(inflation)
unemployment = clean_year(unemployment)
urban = clean_year(urban)
```
